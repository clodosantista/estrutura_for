```python
# 1 -Exercicio executado com for e while
T = int(input("entre com numero inteiro maior que 1"))
print("For:")
for t in range (T):
  print(f"{T-t}", end="")
print("\n")
print("while:")
i=0
while i < T:
  print(f"{T-i}", end="")
  i += 1
```

```python
 # 3  Solicitar o numero inteiro positivo N ao usuário e imprimir i2 para 0 < i < N. Por exemplo, para N= 5, serão impresos:
1
4
9
19

N= int(input("Digite um numero inteiro para N: "))
#FOR
print("For:")
for i in range (1, N):
  print(f"{i*i}")

#while
print("While:")
i = 1
while i < N:
  print(f"{i*i}")
  i += 1
```

```python
# 6 Fazer um programa que solicita um número inteiro positivo N ao usuário e imprime o seguinte:
"""
N= 1:
 -
| |
 -


 N= 2:
 --
|  |
|  |
 --


N= 5:
 -----
|     |
|     |
|     |
|     |
|     |
|     |
 -----

"""

valores_N = [1, 2, 5] #as variaveis que precisa imprimir


for N in valores_N: # fazer a repetição para o N 1
    print(f"N = {N}:")
    print(' ' + '-' * N) #trazer a resposta na tela de um espaço e 1 traço
    for _ in range(N):
        print('|' + ' ' * N + '|') #trazer a resposta na tela com uma barra + um espaço e multiplicar o N e colocar mais uma barra


    # Imprime a linha inferior
    print(' ' + '-' * N)  #preencher um espaço +  a linha inferior e multiplicar o N para trazer a qtd desejada
    print()  # Add uma linha em branco
```
